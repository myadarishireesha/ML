```python
import pandas as pd
#ins=0       #global variable
    
def load_data(path):
    
    ins = pd.read_csv(path)   #local varibale
    
    print("ins --  object created")
    return ins
```

```python
df=load_data("https://raw.githubusercontent.com/rktrojan/DataSciencePython/main/DataFiles/insurance.csv")
#calling the function

df
```

```python
so, here there is no relation between these 2 ins values.
so, inorder to get a relation between the function and ins values we creating class
```

```python
#with class

import pandas as pd

class DL():
    ins=0    #global variable, we can access from outside the function



    def load_data(path):
        ins = pd.read_csv(path)    #local variable, which is access from inside the function
    
        print("ins --  object created")
        return ins    #is to return/print local variable
```

```python
a=DL()    #creating an object
df1=a.load_data("https://raw.githubusercontent.com/rktrojan/DataSciencePython/main/DataFiles/insurance.csv")
df1
```
